```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class ComplementState(TypedDict):
    message : str

def ComplementAgent(state:ComplementState) -> ComplementState:
    state['message'] =  state['message'] + ", you are going an amazing job!!"
    return state
```

```python
graph = StateGraph(ComplementState)

graph.add_node("complement", ComplementAgent)
graph.set_entry_point("complement")
graph.set_finish_point("complement")

app = graph.compile()
```

```python
result = app.invoke({"message":"BOB"})
result["message"]
```
