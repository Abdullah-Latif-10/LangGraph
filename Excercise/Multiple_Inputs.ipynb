```python
from typing import TypedDict,List
from langgraph.graph import StateGraph
import math
```

```python
class AgentState(TypedDict):
    value : List[int]
    name : str
    operation : str
    result : str
```

```python
import math
def processingAgent(state : AgentState) -> AgentState :
    """ This is handle multiple values """
    rs = 0
    if state['operation']=="*":
        rs = math.prod(state['value'])
    else :
        rs = sum(state['value'])
    state['result'] = f"Hi {state['name']} , your answer is: {rs}"

    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("Processor", processingAgent)
graph.set_entry_point("Processor")
graph.set_finish_point("Processor")

app = graph.compile()
```

```python
answer = app.invoke({"value":[1,2,3,4] ,"name":"Jack Sparrow","operation":"*","result":""})
print(answer['result'])
```
